```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://realpython.github.io/fake-jobs/'
request = requests.get(url)
request
```

```python
info = BeautifulSoup(request.content, 'html.parser')
info
```

```python
title = info.find(class_='title is-1')
title.text.strip()
```

```python
div = info.find(id='ResultsContainer')
div.text
```

```python
jobs = info.find_all(class_='column is-half')
jobs
```

```python
for job in jobs:
    print(job.text.strip())
    job_title = info.find(class_='title is-5')
    print(job_title.text.strip())
    sub_title = info.find(class_ = 'subtitle is-6 company')
    print(sub_title.text.strip())
    location =  info.find(class_='location')
    print(location.text.strip())

```

### How find a div in a div

```python
job_title = info.find_all(class_=f"title is-5")
jobs = [job.text for job in job_title]
jobs
```

### Subtitle

```python
subtitle = info.find_all(class_='subtitle is-6 company')
subtitles = [sub.text.strip() for sub in subtitle]
subtitles
```

### Loc & Date

```python
locanddate = info.find_all(class_='content')
locationanddate = [loc.text.strip() for loc in locanddate]
locationanddate
```

```python
description = info.find(class_="subtitle is-3")
description.text.strip()
```

```python
div_data = info.find_all(class_='column is-half')
div_data
```

```python
jobs = info.find_all(class_='column is-half')
jobs
jobs_data =[]
for job in jobs:
    # print(job)
    job_title = job.find(class_='title is-5')
    subtitle = job.find(class_='subtitle is-6 company')
    location = job.find(class_='location')
    date_job = job.find(class_='is-small has-text-grey')
    data = [job_title.text,subtitle.text,location.text,date_job.text]
    jobs_data.append(data)
```
